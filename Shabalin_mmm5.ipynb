```python
import numpy as np
def matr():
    matrix=np.zeros([100,100])
    for i in range (99):
        for j in range(i+2):
                matrix[i][j]=1/(i+1)
                if j==i:
                   matrix[i][j]=10
    for i in range(100):
        matrix[99][i]=1/100
        if i==99:
            matrix[99][i]=10
            
    return matrix

a=matr()
print(a)
def mzn():
    count=0
    mzn=np.zeros([1,100])
    for i in range(100):
        count+=1
        mzn[0][i]=count
    return mzn[0]
b=mzn()
print(b)

Limit=1000
def Yakobi(A,B):
    x = np.zeros(100) 
    for it_count in range(1,Limit): 
        x_new = np.zeros(100) 
        for i in range(A.shape[0]): 
            s1 = np.dot(A[i, :i], x[:i]) 
            s2 = np.dot(A[i, i + 1 :], x[i + 1 :]) 
            x_new[i] = (B[i] - s1 - s2) / A[i, i]
        if np.allclose(x, x_new, rtol=1e-8):
            break
        x = x_new
    for i in range(100):
        print("X{0}\t= \t{1}".format(i+1,x[i]))
    return x
def Gauss(A, B):
    n = len(B)
    for i in range(n):
        a = A[i][i]
        for j in range(i + 1, n):
            stro = A[j][i] / a
            B[j] -= stro * B[i]
            for k in range(i, n):
                A[j][k] -= stro * A[i][k]

    x = [0] * n
    for i in range(n - 1, -1, -1):
        summ = sum(A[i][j] * x[j] for j in range(i + 1, n))
        x[i] = (B[i] - summ) / A[i][i]
    for i in range(100):
        print("X{0}\t= \t{1}".format(i+1,x[i]))
    return x
w,v=np.linalg.eig(a)
la=np.linalg.eig(a)

print("Lamda max= ",max(w))
print("Lamda min= ",min(w))

def normyacobi(a, b, x):
  Ax = np.dot(a, x)
  r = b - Ax
  return np.linalg.norm(r)

def normgauss(a, b, x):
  Ax = np.dot(a, x)
  r = b - Ax
  return np.linalg.norm(r)



print("Решение методом Якоби:")
x_yacobi = Yakobi(a, b)
norm_yacobi = normyacobi(a, b, x_yacobi)
print("Норма вектора невязки: ", norm_yacobi)

print("Решение методом Гаусса:")
x_gauss = Gauss(a, b)
norm_gauss = normgauss(a, b, x_gauss)
print("Норма вектора невязки: ", norm_gauss)
w,v=np.linalg.eig(a)
print("Lamda max= ",max(w))
print("Lamda min= ",min(w))
```
