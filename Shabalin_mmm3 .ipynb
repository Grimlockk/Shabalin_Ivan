```python
import numpy as np
import matplotlib.pyplot as plt
E=10**(-4)

def func(x):
    return (x**(5))+12*(x**(3))-6*(x**(2))-x-10

def dihtom(a,b,func):
    i=0
    c=(a+b)/2
    while(abs(func(c)))>=E:
        c=(a+b)/2
        if(func(a)*func(c))<0:
            a=a
            b=c
        else:
            a=c
            b=b
        i+=1
    print(i)    
    return (a+b)/2



x=np.arange(-9.0,9.0,0.1)
plt.grid()
plt.ylim(-10,10)
plt.xlim(0.0,2.5)
plt.plot(x,func(x))
plt.show()

print(f'Корень, методом дихтомии, = {dihtom(0,2.5,func)}')
```
